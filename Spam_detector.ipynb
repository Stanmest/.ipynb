```
import pandas as pd
```

```
df = pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter01/Dataset/dataset_44_spambase.csv')
```

```
df.head()
```

```
df.describe()
```

```
target = df.pop("class")
```

```
print(target)
```

```
from sklearn.ensemble import RandomForestClassifier
```

```
seed = 168
```

```
rf_model = RandomForestClassifier(random_state=seed)
```

```
rf_model.fit(df,target)
```

```
preds = rf_model.predict(df)
```

```
from sklearn.metrics import accuracy_score
```

```
acc_score = accuracy_score(target, preds)
```

```
print(acc_score)
```
